```python
orders_products_merged_4_7['spending'] = orders_products_merged_4_7.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# step 7
orders_products_merged_4_7['frequency'] = orders_products_merged_4_7.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
orders_products_merged_4_7.loc[orders_products_merged_4_7['frequency'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
orders_products_merged_4_7.loc[(orders_products_merged_4_7['frequency'] <= 20) & (orders_products_merged_4_7['frequency'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
orders_products_merged_4_7.loc[orders_products_merged_4_7['frequency'] <= 10, 'frequency_flag'] = 'Frequent customer'
```
