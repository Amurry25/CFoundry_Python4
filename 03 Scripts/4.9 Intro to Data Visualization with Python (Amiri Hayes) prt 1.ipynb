# 0.4 Data Consistency 

```python
# (1) Checking for missing values
df.describe()
```

```python
df.isnull().sum()
```

The only column/heading with missing values is "first_name". I am hesistent to delete any data becuase it may contain important info that I may need and would prefer to change the data to "str" as most the data is names.

```python
# (2) Checking for duplicated rows
df[df.duplicated()]
```

There appear to be no duplicate rows. 

```python
# (3) Checking for columns with mixed data
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
df['first_name'] = df['first_name'].astype('str')
```

```python
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

Now there are no columns with mixed data.
